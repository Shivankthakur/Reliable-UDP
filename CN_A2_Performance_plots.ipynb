```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pathdir = '/content/drive/MyDrive/##### ONLINE Classes SEM 3-2/Computer Networks/CN A2/'
```

```python
df = pd.read_csv(pathdir+'Loss_%.csv', header=None)
df1 = df.iloc[:, 1:3]
print(df1)

xp = df1.iloc[:,0].tolist()
yp = df1.iloc[:,1].tolist()
plt.title("Throughput vs Loss")
plt.xlabel("Loss%")
plt.ylabel("Throughput(bytes/sec)")
plt.grid()
plt.plot(xp,yp, marker = 'o', mfc = 'r', mec='black')
plt.savefig(pathdir+'loss.png')
plt.show()
```
